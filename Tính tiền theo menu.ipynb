```python
from tkinter import *
from tkinter import ttk
from math import *

def selection():
    total = int(bunbo.get())*int(combobox1.get())+int(pho.get())*int(combobox2.get())+int(lau.get())*int(combobox3.get())
    if var1.get()==1:
        total = total
    else:
        total = total*0.95
    label4 = Label(root, text=f'Tổng bill: {total}').place(x=350, y=182)
    
#     if python == 1:
#         messagebox.showinfo('Python course', 'Bạn đăng ký khóa Python')
#     else:
#         messagebox.showinfo('Python course', 'Bạn không đăng ký khóa Python')
#     messagebox.showinfo('Android course', androidVariable.get())
#     messagebox.showinfo('Angular course', angularVariable.get())

root = Tk()
root.title('Menu')
root.geometry('500x300')

# Chọn món
label1 = Label(root, text='Chọn món').place(x=10, y=10)
bunbo=IntVar()
pho=IntVar()
lau=IntVar()
# pythonVariable.set(1) # cài giá trị mặc định của Python là 1 (checked)
checkbutton1 = Checkbutton(root, text='Bún bò (30k)', onvalue=30000, offvalue=0, variable=bunbo)
checkbutton1.place(x=10, y=30)

checkbutton2 = Checkbutton(root, text='Phở (40k)', onvalue=40000, offvalue=0, variable=pho)
checkbutton2.place(x=10, y=60)

checkbutton3 = Checkbutton(root, text='Lẩu (100k)', onvalue=100000, offvalue=0, variable=lau)
checkbutton3.place(x=10, y=90)


# Chọn số lượng
label2 = Label(root, text='Số lượng').place(x=300,y=10)
values = [f'{i}' for i in range(0,11,1)]
combobox1 = ttk.Combobox(root, values=values)
combobox1.place(x=300, y=30)
combobox1.set(0)

combobox2 = ttk.Combobox(root, values=values)
combobox2.place(x=300, y=60)
combobox2.set(0)

combobox3 = ttk.Combobox(root, values=values)
combobox3.place(x=300, y=90)
combobox3.set(0)

# Hình thức thanh toán
label3 = Label(root, text='Hình thức thanh toán').place(x=50, y=130)
                
var1=IntVar()
var1.set(1)

radiobutton1 = Radiobutton(root, text='Tiền mặt', value=1, variable=var1)
radiobutton1.place(x=50, y=150)

radiobutton2 = Radiobutton(root, text='Chuyển khoản (giảm 5% tổng HĐ)', value=2, variable=var1)
radiobutton2.place(x=250, y=150)

# Thành Tiền button
button = Button(root, text='Thành tiền', command=selection)
button.place(x=250, y=180)
button.bind('<Button>')

root.mainloop()
```
